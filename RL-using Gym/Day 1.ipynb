#### Generating episodes

```python
import gym

env = gym.make("FrozenLake-v1")
num_episodes = 30

num_timesteps = 20


for e in range(num_episodes): # This will generate the number of episodes we want

    print('Time Step 0: ')
    state = env.reset()
    env.render()
    Return = 0
    for t in range(num_timesteps):#This will generate the steps our agent is going to take

        random_action = env.action_space.sample() #We randomly pick an action from our action space 
        next_state, reward, done, info = env.step(random_action) #We pass that action to the step
        Return = Return + reward

        print(f'Time step {t+1}:')
        env.render()
        
        if done:
            break
    print(f'The Total return is: {Return}')
```

##### Computing the optimal value function

* First of all it is essential to compute the optimal value function


$V^*(s) = maxQ^*(s,a)$

```python
# first we will define a function that uses the value iteration method
def value_iteration(env):
    # We set the number of iteration
    num_iterations = 1000
    threshold = 1e-20 #set the threshold
    gamma = 1.0 # set the discount number
    value_table = np.zeros(env.observation_space.n)# set the value of all states to zero
    
    for i in range(num_iterations):# for every iteration
        
        updated_value_table = np.copy(value_table)#update the value
        print(f'this is the updated value: {updated_value_table}')
        for s in range(env.observation_space.n):
            Q_values = [sum([prob*(r+gamma*updated_value_table[s_])
                            for prob, s_, r, _ in env.P[s][a]]) 
                                for a in range(env.action_space.n)]
            print(f'This is the q_value {Q_values}')
            value_table[s] = max(Q_values)
        
        if (np.sum(np.fabs(updated_value_table-value_table)) <= threshold):
            break
        

    return value_table
```

```python
value_iteration(env)
```

####  We want to extract the optimal policy from the optimal value we got 

```python
def extract_policy(value_table):
    gamma = 1.0
    policy = np.zeros(env.observation_space.n)
    for s in range(env.env.observation_space.n):
        Q_values = [sum([prob*(r+gamma*value_table[s_])
                            for prob, s_, r, _ in env.P[s][a]]) 
                                for a in range(env.action_space.n)]
        policy[s] = np.argmax(np.array(Q_values))
    return policy
```

```python
optimal_value_function = value_iteration(env)
```
